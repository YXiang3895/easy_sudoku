```python

def nine_x_nine(row, colum, array):
    remain_list = []
    for i in range (9):
        num = i + 1
        remain_list.append(num)


    smallgrid_row = row // 3
    smallgrid_colum = colum // 3

    row = smallgrid_row *3
    colum = smallgrid_colum *3   
    for i in range(row,row+3,1):   
        for j in range(colum,colum+3,1):
            try:
                remain_list.remove(array[i][j])
            except:
                continue


    count_limit_value = len(remain_list) 
    return count_limit_value,remain_list

def chk_row(row,remain_list,array):

    
    for number in array[row]:     
        try:
            remain_list.remove(number)
        except:
            continue
        
    count_limit_value = len(remain_list)
    return count_limit_value,remain_list

def chk_colum(colum,remain_list,array):

    for row in range(9):
        number = array[row][colum]    
        try:
            remain_list.remove(number)
        except:
            continue
        
    count_limit_value = len(remain_list)
    return count_limit_value,remain_list

def caculate_valid_num(row,colum):
    count, valid_numbers = nine_x_nine(row,colum,sudoku)
    count, valid_numbers = chk_row(row,valid_numbers,sudoku)
    count, valid_numbers = chk_colum(colum,valid_numbers,sudoku)
    return valid_numbers

def repaint_sudoku(times):
    is_done = False
    for row in range(9):
        for colum in range(9):
            if update_matrix[row][colum] == 1:
                try:
                    sudoku[row][colum] = caculate_valid_num(row,colum).pop()
                    update_matrix[row][colum] = 0
                    is_done =  True
                except Exception as e:
                    print(f"error was occured: {e}")
            if is_done:
                break
        if is_done:
            break

    print(f"{times}------------This-is-sudoku----------------")
    for i in range(9):        
        print(sudoku[i])        


def repaint_update_matrix(times):
    for row in range(9):
        for colum in range(9):
            if sudoku[row][colum] != 0:
                # update_matrix[row][colum] = 0
                # count = sudoku[row][colum]
                pass
            else:
                count = 0
                valid_numbers = []
                count, valid_numbers = nine_x_nine(row,colum,sudoku)
                count, valid_numbers = chk_row(row,valid_numbers,sudoku)
                count, valid_numbers = chk_colum(colum,valid_numbers,sudoku)
                update_matrix[row][colum] = count

    print(f"----------This-is-update-table-------------")
    for i in range(9):
        print(update_matrix[i])

def current_status(times):
    print(f"{times}------------This-is-sudoku----------------")
    for i in range(9):
        print(sudoku[i])
    print(f"----------This-is-update-table-------------")
    for i in range(9):
        print(update_matrix[i])

                     # colum
sudoku = [ [0, 5, 1, 0, 7, 9, 0, 4, 0],     
           [9, 4, 0, 0, 0, 2, 0, 0, 0],     
           [0, 6, 7, 0, 3, 1, 0, 0, 0],            
           [5, 0, 9, 0, 0, 8, 0, 0, 1],     #r 
           [4, 1, 0, 0, 0, 0, 0, 3, 6],     #o
           [7, 0, 0, 1, 0, 0, 5, 0, 2],     #w 
           [0, 0, 0, 3, 9, 0, 7, 1, 0], 
           [0, 0, 0, 8, 0, 0, 0, 5, 9], 
           [0, 9, 0, 7, 2, 0, 6, 8, 0]  ]

update_matrix = [[0 for _ in range(9)] for _ in range(9)]
for row in range(9):
    for colum in range(9):
        times_counter += 1
        if sudoku[row][colum] != 0:
            update_matrix[row][colum] = 0
            count = sudoku[row][colum]
            pass
        else:
            count = 0
            count, valid_numbers = nine_x_nine(row,colum,sudoku)
            count, valid_numbers = chk_row(row,valid_numbers,sudoku)
            count, valid_numbers = chk_colum(colum,valid_numbers,sudoku)
            update_matrix[row][colum] = count

def append_to_html(sudoku, update_matrix, html_filename="sudoku_matrices.html"):
    # 轉換矩陣為 HTML 表格格式
    def matrix_to_html(matrix):
        rows = ""
        for row in matrix:
            row_html = ''.join([f"<td>{cell}</td>" for cell in row])
            rows += f"<tr>{row_html}</tr>\n"
        return rows

    # 將 sudoku 和 update_matrix 轉換為 HTML 表格
    sudoku_html = matrix_to_html(sudoku)
    update_matrix_html = matrix_to_html(update_matrix)

    # HTML 內容的模板，包含內嵌 CSS
    html_content = f"""
    <html>
    <head>
        <style>
            .matrix-container {{
                display: flex;
                gap: 50px;
                justify-content: center;
                margin-top: 20px;
            }}

            .matrix {{
                border: 1px solid #000;
                padding: 10px;
            }}

            table {{
                border-collapse: collapse;
            }}

            td {{
                width: 30px;
                height: 30px;
                text-align: center;
                border: 1px solid #ddd;
            }}

            h3 {{
                text-align: center;
            }}
        </style>
    </head>
    <body>
        <div class="matrix-container">
            <div class="matrix">
                <h3>Sudoku Matrix</h3>
                <table>
                    {sudoku_html}
                </table>
            </div>
            <div class="matrix">
                <h3>Update Matrix</h3>
                <table>
                    {update_matrix_html}
                </table>
            </div>
        </div>
    </body>
    </html>
    """

    # 如果 HTML 文件已經存在，則追加內容；如果文件不存在，則創建新文件
    try:
        with open(html_filename, "a") as file:  # 使用 "a" 模式來追加
            file.write(html_content)
            file.write("<br><br>")  # 在兩個更新之間添加空白行
    except Exception as e:
        print(f"Error writing to file: {e}")


current_status(0)
i = 1
is_cont = True
append_to_html(sudoku, update_matrix)

while is_cont :
    sum = 0
    for row in range(9):
        for num in update_matrix[row]:
            sum += num
        if sum > 0:
            break
    if sum == 0:
        is_cont = False
        
    repaint_sudoku(i)
    repaint_update_matrix(i)
    append_to_html(sudoku, update_matrix)

    i += 1
```
